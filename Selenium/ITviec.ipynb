```python
job_no=0
page=0
url_content=[]
url="https://itviec.com/"
while True:
    response = requests.get(url)
    data = response.text
    soup = bs(data,'html.parser')
  
    link_jobs=soup.find_all("div",{"data-controller":"search--job-selection"})
    for link in link_jobs:
        links=link["data-search--job-selection-job-url"]
        print(links)
        url_content.append(links)
        job_no+=1

    url_tag=soup.find('a',{'rel':'next'})
    if url_tag["href"]: 
        url="https://itviec.com/"+ url_tag["href"]
        page+=1
    else:
        break

```

```python
lst_title=[]
lst_company=[]
lst_company_origin=[]
lst_company_type=[]
lst_skill=[]
lst_company_adress=[]
lst_company_employee=[]
url="https://itviec.com/"
for i in url_content:
    job_link=url+i
    response = requests.get(job_link)
    data = response.text
    soup = bs(data,'html.parser')
    job_title= soup.find("h1",{"class":"job-details__title"}).text
    lst_title.append(job_title)
    company=soup.find("div",{"class":"search-page-employer-overview__name"}).text
    lst_company.append(company)
    b=soup.findAll("div",{"class":"svg-icon__text"})
    if len(b)== 7:
        lst_company_origin.append(b[5].text)
        lst_company_type.append(b[2].text)
        lst_company_employee.append(b[3].text)
    elif len(b)==9:
        lst_company_origin.append(b[7].text)
        lst_company_type.append(b[4].text)
        lst_company_employee.append(b[5].text)
    elif len(b)==8:
        lst_company_origin.append(b[6].text)
        lst_company_type.append(b[3].text)
        lst_company_employee.append(b[4].text)
    a=soup.findAll("span")
    lst=tuple([a[i].text.strip() for i in range(3)])
    lst_skill.append(lst)
    lst_company_adress.append(a[3].text)
```
