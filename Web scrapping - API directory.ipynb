Your Python code should scrape the following details from each table row:

• API Name

• API (absolute) URL

• API Category

• API Description


```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
number=0
api_dic={}
url="https://www.programmableweb.com/category/all/apis"
while True:
    respone=requests.get(url)
    data=respone.text
    soup=bs(data,"html.parser")
    apis=soup.find_all("tr")
    
    for api in apis:
        api_names_tag=api.find("a")
        api_names=api_names_tag.get_text() if api_names_tag else "N/A"
        api_link_tag=api.find("a")
        api_link=api_link_tag.get("href") if api_link_tag else "N/A"
        api_category_tag=api.find("td",{"class":"views-field views-field-field-article-primary-category"})
        api_category=api_category_tag.text if api_category_tag else "N/A"
        api_des_tag=api.find("td",{"class":"views-field views-field-search-api-excerpt views-field-field-api-description hidden-xs visible-md visible-sm col-md-8"})
        api_des=api_des_tag.text if api_des_tag else "N/A"
        #rint(api_names,"\n",api_link,"\n",api_category,"\n",api_des,"\n------")
        number+=1
        api_dic[number]=[api_names,api_link,api_category,api_des]


    url_tag=soup.find('a',{'title':'Go to next page'})
    if url_tag.get('href'):
        url="https://programmableweb.com"+url_tag.get('href')
        print(url)
    else:
        break
            
print("Total API: ",number)
api_dc=pd.DataFrame.from_dict(api_dic,orient ="index",columns=["Name","Link","Category","Description"])

#preview the first ten rows
api_dc.head()                                       


# extract into CSV file
api_dc.to_csv("api.csv")
```
