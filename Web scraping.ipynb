```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://boston.craigslist.org/search/sof"
response = requests.get(url)
#response 200 means the url works well and be right direction
response
```

```python
#print(response)
data = response.text
data
```

```python
#print(data)
soup = BeautifulSoup(data,'html.parser')
```

```python
tags = soup.find_all('a')

for tag in tags:
    print(tag.get('href'))
```

```python
for dataid in tags:
    print(dataid.get('data-id'))
```

```python
titles= soup.find_all("a",{"class":"result-title"})
```

```python
for title in titles:
    print(title.text)
```

```python
address= soup.find_all("span",{"class":"result-hood"})
for add in address:
    print(add.text)
```

```python
jobs= soup.find_all("div",{"class":"result-info"})
```

```python
for job in jobs:
    title= job.find("a",{"class":"result-title"}).text
    location_tag=job.find("span",{"class":"result-hood"})
    location=location_tag.text if location_tag else "N/A"
    time=job.find("time",{"class":"result-date"}).text
    link=job.find("a",{"class":"result-title hdrlnk"}).get("href")
    job_response=requests.get(link)
    job_data=job_response.text
    job_soup=BeautifulSoup(job_data,"html.parser")
    job_description_tag=job_soup.find("section",{"id":"postingbody"})
    job_description=job_description_tag.text if job_description_tag else "N/A"
    job_attr_tag=job_soup.find("p",{"class":"attrgroup"})
    job_attr=job_attr_tag.text if job_attr_tag else "N/A"
    print(title,"\n",location,"\n",time,"\n",link,"\n",job_description,"\n",job_attr,"\n------")
```
