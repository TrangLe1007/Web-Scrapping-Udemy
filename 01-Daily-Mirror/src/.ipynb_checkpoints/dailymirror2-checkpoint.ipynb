```python
# web trên bị sao á, em thay bằng Tuổi Trẻ r cũng scrape lấy title nha
import requests
from bs4 import BeautifulSoup as bs
response = requests.get('https://tuoitre.vn/')
print('Visited URL: {}'.format(response.url))
print(response.status_code)
```

```python
soup = bs(response.text, 'html.parser')

type(soup)
```

```python
print(soup.prettify())
```

```python
data_containers = soup.find_all('ul', class_='list-news-content')
print(data_containers)
```

```python
titles = soup.find_all('h3', class_='title-news')
print(titles)
```

```python
for title in titles:
    tenbai = title.find('h3', class_='title-news')
    print(tenbai)
```
