```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
stock = input("Hãy nhập mã công ty bạn muốn xem:")
url = "https://s.cafef.vn/Lich-su-giao-dich-{}-1.chn".format(stock)

r = requests.get(url)
r.content
content = bs(r.content, 'lxml')
x = content.find_all("td", {"class": "Item_DateItem"})
y = content.find_all("td", {"class": "Item_Price10"})
date = []
price =[]

for e in x:
    date.append(e.text)
    
for a in y:
    price.append(a.text.replace('\xa0',""))

for i in range (len(date)):
    print('Giá ngày đóng cửa {} là {}'.format(date[i], price[i*11]))


while 1:
    ans=input(f"\nBạn có muốn tìm mã khác? yes or no.\n")
    if ans != 'yes':
         break
    stock = input("Hãy nhập mã công ty bạn muốn xem:")
    url = "https://s.cafef.vn/Lich-su-giao-dich-{}-1.chn".format(stock)
    r = requests.get(url)
    r.content
    content = bs(r.content, 'lxml')
    x = content.find_all("td", {"class": "Item_DateItem"})
    y = content.find_all("td", {"class": "Item_Price10"})
    date = []
    price =[]

    for e in x:
        date.append(e.text)

    for a in y:
        price.append(a.text.replace('\xa0',""))

    for i in range (len(date)):
        print('Giá ngày đóng cửa {} là {}'.format(date[i], price[i*11]))

```
