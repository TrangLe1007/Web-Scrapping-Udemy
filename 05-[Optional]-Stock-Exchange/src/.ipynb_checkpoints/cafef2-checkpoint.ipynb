```python
url = "https://s.cafef.vn/Lich-su-giao-dich-FPT-1.chn"
driver = webdriver.Edge("D:\DA\msedgedriver")

driver.get(url)
driver.maximize_window()

r = requests.get(url)
r.content

content = bs(r.content, 'lxml')

x = content.find_all("td", {"class": "Item_DateItem"})
y = content.find_all("td", {"class": "Item_Price10"})

date = []
price =[]

for e in x:
    date.append(e.text)
    ee=e.text
for a in y:
    price.append(a.text)
    aa=a.text
    print("Giá ngày", ee, "là", aa)

driver.quit()
```
